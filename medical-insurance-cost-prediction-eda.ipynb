### **Data Preprocessing üêøÔ∏è**

#### **Data Nomenclature üêò**
---------------------------------
##### **Sex Column** üèà
* male = 0
* female = 1

##### **Smoker Column** üéø
* yes = 0
* no = 1

##### **Region Column** üïØÔ∏è
* southeast = 0
* southwest = 1
* northeast = 2
* northwest = 3

```python
# Encoding "sex" column üè°
df.replace({'sex': {'male' : 0, 'female' : 1}}, inplace = True)
# Encoding "smoker" column ‚òòÔ∏è
df.replace({'smoker': {'yes' : 0, 'no' : 1}}, inplace = True)
# Encoding "region" column ‚òÄÔ∏è
df.replace({'region': {'southeast' : 0, 'southwest' : 1, 'northeast' : 2, 'northwest' : 3}}, inplace = True)
```

### **Recognizing Features & Targets** ü¶â

```python
# Splitting features & targets üå±
# Feature Data 
x = df.iloc[:,0:6]
# Target Data 
y = df.iloc[:,-1] 
```

### **Data Splitting** ü¶Ö

```python
# Splitting data into train and test datasets üçÇ
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

### **Model Training & Predition ü¶ö**

#### **Model Training** üöÄ

```python
# Training data on Linear Regression Algorithm üõçÔ∏è
lr = LinearRegression()
# Model Training üìû
lr.fit(X_train, Y_train)
```

#### **Model Evaluation** üåÑ

```python
# Model prediction üíé
pred = lr.predict(X_test)
# R Square value to check credibility of the model üöß
r2Value = r2_score(Y_test,pred)
print("r2 score value : ",r2Value)
```

#### **Observations** ü¶ï
* Model is working well as r2_score is near to 1.

### **Building a Predictive System** üêã

```python
class Pred:
    
    def __init__ (self):
        self.age = None
        self.sex = None
        self.bmi = None
        self.children = None
        self.smoker = None
        self.region = None
        
    def model(self, age, sex, bmi, children, smoker, region):
        
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.region = region
        
        data = np.array([self.age, self.sex, self.bmi, self.children, self.smoker, self.region])
        data = data.reshape(1,-1)
        predCost = lr.predict(data)
        
        return predCost
    
    
if __name__ == '__main__':
        obj = Pred()
        r = obj.model(37,0,22,3,0,3)
        print(r)
        
```
